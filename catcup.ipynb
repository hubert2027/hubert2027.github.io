## Q1: Algorithm Design – University Registration System

```python
# Q1a & Q1d: Registration function with constraint validation
def register_student(student_id, name, course, registered, capacity):
    if not student_id or not name or not course:      # missing data check
        return "Error: Missing data"
    if any(s['id'] == student_id for s in registered):  # duplicate check
        return "Error: Duplicate student"
    if len(registered) >= capacity:                   # capacity check
        return "Error: Course full"
    registered.append({'id': student_id, 'name': name, 'course': course})
    return f"{name} registered for {course}"

registered = []
print(register_student("S001", "Alice", "CS101", registered, 30))  # success
print(register_student("S001", "Alice", "CS101", registered, 30))  # duplicate
print(register_student("", "Bob", "CS101", registered, 30))        # missing data
```

```python
# Q1b: Course allocation using conditionals
def allocate_course(marks):
    if marks >= 75:
        return "Advanced Course"
    elif marks >= 60:
        return "Intermediate Course"
    elif marks >= 50:
        return "Basic Course"
    else:
        return "Remedial Course"

print(allocate_course(85))  # Advanced Course
print(allocate_course(55))  # Basic Course
print(allocate_course(40))  # Remedial Course
```

## Q3: Graph-Based Social Network Analysis

```python
# Q3a: Adjacency list representation
graph = {
    'Alice':   ['Bob', 'Charlie'],
    'Bob':     ['Alice', 'Diana'],
    'Charlie': ['Alice'],
    'Diana':   ['Bob']
}
print("Graph:", graph)
```

```python
# Q3b: BFS traversal
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    result = []
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            result.append(node)
            queue.extend(graph[node])
    return result

print("BFS from Alice:", bfs(graph, 'Alice'))
```

```python
# Q3c: Find mutual friends using sets
def mutual_friends(graph, user1, user2):
    return set(graph[user1]) & set(graph[user2])

print("Mutual friends (Alice, Diana):", mutual_friends(graph, 'Alice', 'Diana'))
print("Mutual friends (Bob, Charlie):", mutual_friends(graph, 'Bob', 'Charlie'))
```

```python
# Q3d: Display full graph structure
def display_graph(graph):
    for user, friends in graph.items():
        print(f"{user} --> {friends}")

display_graph(graph)
```

## Q4: Search Algorithms and Complexity Analysis

```python
# Q4a: Linear and Binary Search
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:   return mid
        elif arr[mid] < target:  low = mid + 1
        else:                    high = mid - 1
    return -1

arr = list(range(100))
print("Linear search for 42:", linear_search(arr, 42))
print("Binary search for 42:", binary_search(arr, 42))
```

```python
# Q4b: Measure execution time
import time

arr = list(range(10000))
target = 9999

t1 = time.time()
linear_search(arr, target)
print(f"Linear time: {time.time() - t1:.6f}s")

t1 = time.time()
binary_search(arr, target)
print(f"Binary time: {time.time() - t1:.6f}s")
```

```python
# Q4c: Big-O comparison with examples
print("O(1)      - Constant:     dict lookup  -> d['key']")
print("O(log n)  - Logarithmic:  binary search -> halves each step")
print("O(n)      - Linear:       linear search -> checks each item")
print("O(n^2)    - Quadratic:    bubble sort   -> nested loops")

# Demonstrate O(1)
d = {i: i*2 for i in range(10000)}
t1 = time.time(); d[9999]; print(f"O(1) time: {time.time()-t1:.8f}s")
```

```python
# Q4d: Plot execution time
import matplotlib.pyplot as plt

sizes = [100, 500, 1000, 5000, 10000]
linear_times = []
binary_times = []

for n in sizes:
    arr = list(range(n))
    target = n - 1

    t1 = time.time()
    linear_search(arr, target)
    linear_times.append(time.time() - t1)

    t1 = time.time()
    binary_search(arr, target)
    binary_times.append(time.time() - t1)

plt.plot(sizes, linear_times, label='Linear O(n)', marker='o')
plt.plot(sizes, binary_times, label='Binary O(log n)', marker='s')
plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Search Algorithm Execution Time')
plt.legend()
plt.show()
```

## Q5: Algorithm Design Strategies for Logistics

```python
# Q5a: Greedy algorithm – pick shortest route first
def greedy_route(routes):
    return sorted(routes, key=lambda x: x[1])

routes = [('Kigali', 10), ('Butare', 30), ('Gisenyi', 20)]
print("Greedy route order:", greedy_route(routes))
```

```python
# Q5b: Dynamic programming – 0/1 Knapsack (cost optimization)
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i-1][w], values[i-1] + dp[i-1][w - weights[i-1]])
            else:
                dp[i][w] = dp[i-1][w]
    return dp[n][capacity]

print("Max value:", knapsack([2, 3, 4], [3, 4, 5], 5))  # 7
```

```python
# Q5c: Backtracking – try all route permutations
def backtrack(cities, path=[]):
    if len(path) == len(cities):
        return [path[:]]
    result = []
    for city in cities:
        if city not in path:
            path.append(city)
            result += backtrack(cities, path)
            path.pop()
    return result

all_routes = backtrack(['A', 'B', 'C'])
print("All possible routes:", all_routes)
```

```python
# Q5d: Compare strategies
print("Strategy      | Speed       | Optimal?")
print("Greedy        | O(n log n)  | Sometimes")
print("Dynamic Prog  | O(n x W)    | Always")
print("Backtracking  | O(n!)       | Always (exhaustive)")
```

## Q6: Sorting Algorithms for Academic Records

```python
# Q6a: Bubble Sort
def bubble_sort(arr):
    arr = arr[:]
    for i in range(len(arr)):
        for j in range(len(arr) - i - 1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

print("Bubble sort:", bubble_sort([64, 25, 12, 22, 11]))
```

```python
# Q6b: Selection Sort
def selection_sort(arr):
    arr = arr[:]
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

print("Selection sort:", selection_sort([64, 25, 12, 22, 11]))
```

```python
# Q6c: Merge Sort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]: result.append(left[i]); i += 1
        else:                   result.append(right[j]); j += 1
    return result + left[i:] + right[j:]

print("Merge sort:", merge_sort([64, 25, 12, 22, 11]))
```

```python
# Q6d: Compare execution times
import time

arr = list(range(1000, 0, -1))  # worst case: reversed array

t1 = time.time(); bubble_sort(arr);    print(f"Bubble:    {time.time()-t1:.4f}s  O(n^2)")
t1 = time.time(); selection_sort(arr); print(f"Selection: {time.time()-t1:.4f}s  O(n^2)")
t1 = time.time(); merge_sort(arr);     print(f"Merge:     {time.time()-t1:.4f}s  O(n log n)")
```

## Q7: Hashing and Tree Data Structures

```python
# Q7a & Q7b: Hash Table with chaining (collision handling)
class HashTable:
    def __init__(self, size=10):
        self.table = [[] for _ in range(size)]
        self.size = size

    def put(self, key, value):
        idx = hash(key) % self.size
        self.table[idx].append((key, value))   # chaining = append to list

    def get(self, key):
        idx = hash(key) % self.size
        for k, v in self.table[idx]:
            if k == key:
                return v
        return None

ht = HashTable()
ht.put('ACC001', 'Alice')
ht.put('ACC002', 'Bob')
print("Search ACC001:", ht.get('ACC001'))
print("Search ACC002:", ht.get('ACC002'))
print("Collision handling: multiple keys stored as list in same bucket")
```

```python
# Q7c: Binary Search Tree (BST)
class Node:
    def __init__(self, val):
        self.val = val
        self.left = self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, val):
        def _ins(node, val):
            if not node: return Node(val)
            if val < node.val: node.left  = _ins(node.left, val)
            else:              node.right = _ins(node.right, val)
            return node
        self.root = _ins(self.root, val)

    def search(self, val):
        node = self.root
        while node:
            if val == node.val:   return True
            elif val < node.val:  node = node.left
            else:                 node = node.right
        return False

bst = BST()
for v in [50, 30, 70, 20, 40]:
    bst.insert(v)
print("Search 40:", bst.search(40))   # True
print("Search 99:", bst.search(99))   # False
```

## Q8: Graph Algorithms for Transport Systems

```python
# Q8c: Build graph using adjacency list
road_graph = {
    'Kigali':    [('Butare', 50), ('Ruhengeri', 80)],
    'Butare':    [('Kigali', 50), ('Gisenyi', 60)],
    'Ruhengeri': [('Kigali', 80)],
    'Gisenyi':   [('Butare', 60)]
}
print("Graph built:", road_graph)
```

```python
# Q8a: BFS and DFS traversal
from collections import deque

simple_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['B']
}

def bfs(graph, start):
    visited, queue, result = set(), deque([start]), []
    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node); result.append(node)
            queue.extend(graph[node])
    return result

def dfs(graph, node, visited=None):
    if visited is None: visited = set()
    visited.add(node)
    result = [node]
    for neighbour in graph[node]:
        if neighbour not in visited:
            result += dfs(graph, neighbour, visited)
    return result

print("BFS:", bfs(simple_graph, 'A'))
print("DFS:", dfs(simple_graph, 'A'))
```

```python
# Q8b: Dijkstra's shortest path
import heapq

def dijkstra(graph, start):
    dist = {node: float('inf') for node in graph}
    dist[start] = 0
    heap = [(0, start)]
    while heap:
        d, u = heapq.heappop(heap)
        for v, w in graph[u]:
            if d + w < dist[v]:
                dist[v] = d + w
                heapq.heappush(heap, (dist[v], v))
    return dist

road = {
    'A': [('B', 4), ('C', 2)],
    'B': [('D', 3)],
    'C': [('B', 1), ('D', 5)],
    'D': []
}
print("Shortest distances from A:", dijkstra(road, 'A'))
```

```python
# Q8d: Minimum Spanning Tree – Kruskal's algorithm
def kruskal(edges, nodes):
    edges.sort(key=lambda x: x[2])      # sort by weight
    parent = {n: n for n in nodes}

    def find(n):
        while parent[n] != n:
            n = parent[n]
        return n

    mst = []
    for u, v, w in edges:
        if find(u) != find(v):           # no cycle
            mst.append((u, v, w))
            parent[find(u)] = find(v)
    return mst

edges = [('A','B',4), ('A','C',2), ('C','B',1), ('B','D',3)]
nodes = ['A', 'B', 'C', 'D']
print("MST:", kruskal(edges, nodes))
```

## Q9: String Matching in Search Engines

```python
# Q9a: Naive string matching
def naive_search(text, pattern):
    n, m = len(text), len(pattern)
    found = []
    for i in range(n - m + 1):
        if text[i:i+m] == pattern:
            found.append(i)
    return found

print("Naive search:", naive_search("AABAACAADAA", "AAD"))   # [6]
```

```python
# Q9b: KMP algorithm
def kmp_search(text, pattern):
    n, m = len(text), len(pattern)

    # Build lps (failure) table
    lps = [0] * m
    j = 0
    for i in range(1, m):
        while j > 0 and pattern[i] != pattern[j]: j = lps[j-1]
        if pattern[i] == pattern[j]: j += 1
        lps[i] = j

    # Search
    j = 0
    found = []
    for i in range(n):
        while j > 0 and text[i] != pattern[j]: j = lps[j-1]
        if text[i] == pattern[j]: j += 1
        if j == m:
            found.append(i - m + 1)
            j = lps[j-1]
    return found

print("KMP search:", kmp_search("AABABACABA", "ABA"))   # [1, 3, 7]
```

```python
# Q9c & Q9d: Compare algorithms
import time

text    = "AABAACAADAA" * 1000
pattern = "AAD"

t1 = time.time(); naive_search(text, pattern); print(f"Naive: {time.time()-t1:.6f}s  O(n x m)")
t1 = time.time(); kmp_search(text, pattern);   print(f"KMP:   {time.time()-t1:.6f}s  O(n + m)")
print("KMP is faster because it skips re-checking using the lps table")
```

## Q10: Emerging Algorithms & Complexity Problems

```python
# Q10a: Knapsack problem (Dynamic Programming)
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for w in range(capacity + 1):
            if weights[i-1] <= w:
                dp[i][w] = max(dp[i-1][w], values[i-1] + dp[i-1][w - weights[i-1]])
            else:
                dp[i][w] = dp[i-1][w]
    return dp[n][capacity]

weights  = [1, 3, 4, 5]
values   = [1, 4, 5, 7]
capacity = 7
print("Max value (Knapsack):", knapsack(weights, values, capacity))  # 9
```

```python
# Q10b: Traveling Salesman Problem (brute force / exact)
from itertools import permutations

def tsp(dist, start):
    cities = [c for c in dist if c != start]
    best_cost  = float('inf')
    best_route = None
    for perm in permutations(cities):
        route = [start] + list(perm) + [start]
        cost = sum(dist[route[i]][route[i+1]] for i in range(len(route)-1))
        if cost < best_cost:
            best_cost  = cost
            best_route = route
    return best_route, best_cost

dist = {
    'A': {'A':0, 'B':10, 'C':15, 'D':20},
    'B': {'A':10,'B':0,  'C':35, 'D':25},
    'C': {'A':15,'B':35, 'C':0,  'D':30},
    'D': {'A':20,'B':25, 'C':30, 'D':0 }
}
route, cost = tsp(dist, 'A')
print("Best route:", route)
print("Best cost:", cost)
```

```python
# Q10c: Compare exact vs approximation
print("Method          | Quality       | Speed    | Use when")
print("Exact (DP/BF)   | Perfect       | O(n!)    | Small inputs (n < 15)")
print("Greedy (approx) | Good enough   | O(n^2)   | Large inputs")

# Greedy TSP approximation (nearest neighbour)
def greedy_tsp(dist, start):
    unvisited = [c for c in dist if c != start]
    route = [start]
    current = start
    while unvisited:
        nearest = min(unvisited, key=lambda c: dist[current][c])
        route.append(nearest)
        current = nearest
        unvisited.remove(nearest)
    route.append(start)
    cost = sum(dist[route[i]][route[i+1]] for i in range(len(route)-1))
    return route, cost

route, cost = greedy_tsp(dist, 'A')
print("Greedy route:", route, "| Cost:", cost)
```
